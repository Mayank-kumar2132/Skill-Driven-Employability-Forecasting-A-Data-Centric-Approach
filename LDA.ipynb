```python
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df0 = pd.read_csv("DATA/megaData.csv")
df0
```

```python
class_counts = df0['Employed'].value_counts()
class_counts
```

```python
X=df0.drop("Employed", axis=1)
X
```

```python
y = df0['Employed']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
X.shape
```

```python
lda = LinearDiscriminantAnalysis(n_components = 1)
```

```python
X_train_lda = lda.fit_transform(X_train, y_train)
```

```python
X_test_lda = lda.transform(X_test)
```

# Now using LDA Train and Test data for different models

### LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
classifier_log = LogisticRegression()
```

```python
classifier_log.fit(X_train_lda, y_train)
```

```python
y_pred_log = classifier_log.predict(X_test_lda)
```

```python
class_report_log = classification_report(y_test, y_pred_log)
print(class_report_log)
```

### DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
classifier_dt = DecisionTreeClassifier()
```

```python
classifier_dt.fit(X_train_lda, y_train)
```

```python
y_pred_dt = classifier_dt.predict(X_test_lda)
```

```python
class_report_dt = classification_report(y_test, y_pred_dt)
print(class_report_dt)
```

### GRADIENT BOOSTING CLASSIFIER (GBT)

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report
```

```python
clf_gbt = GradientBoostingClassifier(
    loss='log_loss',  # Use 'log_loss' for probabilistic outputs
    learning_rate=0.1,  # Shrinkage factor
    n_estimators=100,  # Number of boosting stages
    max_depth=5,  # Maximum depth of individual trees
    subsample=1.0,  # Fraction of samples used for fitting base learners
    random_state=42  # Random seed for reproducibility
)
```

```python
clf_gbt.fit(X_train_lda, y_train)
```

```python
y_pred_gbt = clf_gbt.predict(X_test_lda)
```

```python
class_report_gbt = classification_report(y_test, y_pred_gbt)
print(class_report_gbt)
```
